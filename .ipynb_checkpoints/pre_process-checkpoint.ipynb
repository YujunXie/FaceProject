# 人脸图像预处理

通过灰度变换，上下采样，幂律变换，滤波，直方图均衡化，归一化等操作提高人脸识别的准确率。

### 导入图像

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image, ImageFilter

%matplotlib inline

path = 'Detection/data/FaceImage/xzq.png'
im = Image.open(path)

plt.axis('off')
plt.imshow(im)
```

### 灰度变换

```python
gray_im = im.convert('L')

plt.figure(num='gray_faces',figsize=(8,8)) 

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('gray image')
plt.imshow(gray_im)
plt.axis('off')
```

### 采样

高斯金字塔：将原图像作为最底层图像G0，利用高斯核（5*5）对其进行卷积，然后对卷积后的图像进行下采样（去除偶数行和列）得到上一层图像G1，将此图像作为输入，重复卷积和下采样操作得到更上一层图像，反复迭代多次，形成一个金字塔形的图像数据结构，即高斯金字塔。
高斯金字塔上采样：图像放大，分辨率上升
高斯金字塔上采样：图像缩小，分辨率降低

拉普拉斯金字塔：用于重建图像，也就是预测残差，对图像进行最大程度的还原。比如一幅小图像重建为一幅大图，原理：用高斯金字塔的每一层图像减去其上一层图像上采样并高斯卷积之后的预测图像，得到一系列的差值图像即为 LP 分解图像。
拉普拉斯金字塔：还原上下采样丢失的信息。

```python
im = np.asarray(im)

#upsample
up_im = cv2.pyrDown(im)

#downsample
down_im = cv2.pyrUp(im)

# 高斯金字塔转拉普拉斯金字塔
# Li=Gi−PyrUp(Gi+1)
temp_im = cv2.pyrDown(up_im)
temp = cv2.pyrUp(temp_im)
lap_im = up_im - temp

plt.figure(num='faces',figsize=(8,8)) 

plt.subplot(1,4,1)
plt.title('origin image')
plt.imshow(im)

plt.subplot(1,4,2)
plt.title('upsample')
plt.imshow(up_im)

plt.subplot(1,4,3)
plt.title('downsample')
plt.imshow(down_im)

plt.subplot(1,4,4)
plt.title('laplase')
plt.imshow(lap_im)

plt.show()
```

### 幂律变换

基于Gamma灰度矫正，通过对像素的非线性操作，对图像进行光照预处理。
gamma > 1： 处理过亮的图片，进行灰度级压缩
gamma < 1： 处理过暗的图片，对比度增强，使得细节看的更加清楚

```python
im = np.asarray(im)
gamma = 2.2
gamma_im = ((im/255) ** (1/gamma))

plt.figure(num='gamma', figsize = (8,8))

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('gamma image')
plt.imshow(gamma_im)
plt.axis('off')
```

### 滤波

通过滤波对图像进行去噪，平滑，锐化等效果。

```python
plt.figure(num='filter', figsize = (10,10))

plt.subplot(2,4,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

# 细节滤波
# 细节增强滤波,使图像细节更加明显。
# 0, -1,  0,
# -1, 10, -1,
# 0, -1,  0
detail_im = im.filter(ImageFilter.DETAIL)
plt.subplot(2,4,2)
plt.title('detail filter')
plt.imshow(detail_im)
plt.axis('off')

# 平滑滤波
# 突出图像的宽大区域、低频成分、主干部分或抑制图像噪声和干扰高频成分,使图像亮度平缓渐变,减小突变梯度,改善图像质量。
# 1,  1,  1,
# 1,  5,  1,
# 1,  1,  1
smooth_im = im.filter(ImageFilter.SMOOTH)
plt.subplot(2,4,3)
plt.title('smooth filter')
plt.imshow(smooth_im)
plt.axis('off')

# 平滑滤波加强
# 更加平滑自然，但是处理速度明显变慢 
# 1,  1,  1,  1,  1,
# 1,  5,  5,  5,  1,
# 1,  5, 44,  5,  1,
# 1,  5,  5,  5,  1,
# 1,  1,  1,  1,  1
smooth_m_im = im.filter(ImageFilter.SMOOTH_MORE)
plt.subplot(2,4,4)
plt.title('smooth more filter')
plt.imshow(smooth_m_im)
plt.axis('off')

# 锐化滤波
# 补偿图像的轮廓,增强图像的边缘及灰度跳变的部分,使图像变得清晰
# -2, -2, -2,
# -2, 32, -2,
# -2, -2, -2
sharpen_im = im.filter(ImageFilter.SHARPEN)
plt.subplot(2,4,5)
plt.title('sharpen filter')
plt.imshow(sharpen_im)
plt.axis('off')

# 中值滤波
# 有效消除孤立噪声点
median_im = im.filter(ImageFilter.MedianFilter(3))
plt.subplot(2,4,6)
plt.title('median filter')
plt.imshow(median_im)
plt.axis('off')

# 高斯滤波
gaussian_im = im.filter(ImageFilter.GaussianBlur(radius=2))
plt.subplot(2,4,7)
plt.title('gaussian filter')
plt.imshow(gaussian_im)
plt.axis('off')
```

### 直方图均衡化📊

增强局部或者整体的对比度。使人脸图像具有相同的光照条件。

```python
from skimage import exposure

im = np.asarray(im)
plt.figure(num='hist', figsize = (8,8))

plt.subplot(2,2,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

plt.subplot(2,2,2)
plt.title('origin hist image')
plt.hist(im.flatten(), 256)
plt.axis('off')

equ_im = exposure.equalize_hist(im)

plt.subplot(2,2,3)
plt.title('hist equalized image')
plt.imshow(equ_im)
plt.axis('off')

plt.subplot(2,2,4)
plt.title('equalized hist image')
plt.hist(equ_im.flatten(), 256)
plt.axis('off')
```

### 归一化

灰度归一化：将图像尺寸和像素规定在一个范围内，增加图像的对比度，进行光照补偿。(x-min)/(max-min)
几何归一化：人脸对齐

```python
# 灰度归一化
im = Image.open(path)
im = np.asarray(im)

plt.figure(num='normalize', figsize = (8,8))

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

# 像素归一化
for i in im:
    i = (i - np.min(im))/(np.max(im) - np.min(im))

im = Image.fromarray(im)
(x,y) = im.size
print(x,y)

# 尺寸归一化
im = im.resize((250,250))
(x,y) = im.size
print(x,y)

plt.subplot(1,2,2)
plt.title('normalized image')
plt.imshow(im)
plt.axis('off')
```

### 人脸对齐


```python

import math

def transformation_from_points(points1, points2):
    points1 = points1.astype(np.float64)
    points2 = points2.astype(np.float64)
    c1 = np.mean(points1, axis=0)
    c2 = np.mean(points2, axis=0)
    points1 -= c1
    points2 -= c2
    s1 = np.std(points1)
    s2 = np.std(points2)
    points1 /= s1
    points2 /= s2
    U, S, Vt = np.linalg.svd(points1.T * points2)
    R = (U * Vt).T
    return np.vstack([np.hstack(((s2 / s1) * R,c2.T - (s2 / s1) * R * c1.T)), np.matrix([0., 0., 1.])])

path = 'Detection/data/FaceImage/xzq.png'
im = Image.open(path)

coord5points = [[0.31556875000000000, 0.4615741071428571],
 [0.68262291666666670, 0.4615741071428571],
 [0.50026249999999990, 0.6405053571428571],
 [0.34947187500000004, 0.8246919642857142],
 [0.65343645833333330, 0.8246919642857142]]

#乘以图像中人脸宽高
for coord in coord5points:
    coord[0] = coord[0]*96
    coord[1] = coord[1]*112

facepoints = [[254, 163],
              [335, 182],
              [285, 216],
              [276, 208],
              [311, 237]]

'''
coord5points = [[0.31556875000000000, 0.4615741071428571],
 [0.68262291666666670, 0.4615741071428571],
 [0.50026249999999990, 0.6405053571428571],
 [0.34947187500000004, 0.8246919642857142],
 [0.65343645833333330, 0.8246919642857142]]

#阿拉伯
facepoints = [[149, 110],
              [135, 112],
              [97, 121],
              [111, 116],
              [129, 136]]
#改造后
facepoints = [[97, 121],
              [149, 110],
              [129, 136],
              [111, 146],
              [135, 152]]

#薛之谦
#原blog
facepoints = [[259, 137],
              [319, 150],
              [284, 177],
              [253, 206],
              [297, 216]]
#landmark           
facepoints = [[335, 182],
              [311, 177],
              [254, 163],
              [276, 168],
              [285, 216]]
# 改造后 
facepoints = [[254, 163],
              [335, 182],
              [285, 216],
              [276, 208],
              [311, 237]]
'''

# pos1,pos2是变换前后的对应位置关系,输出仿射矩阵M
pos1 = np.float64(np.matrix([[point[0], point[1]] for point in facepoints]))
pos2 = np.float64(np.matrix([[point[0], point[1]] for point in coord5points]))
M = transformation_from_points(pos1,pos2)

im = np.asarray(im)

# 根据旋转矩阵进行仿射变换(原始图像、变换矩阵、人脸图像尺寸)
res = cv2.warpAffine(im, M[:2], (96, 112));

plt.imshow(res)

'''
import dlib

path = 'Detection/data/FaceImage/Abdullah_Gul_0004.jpg'
im = Image.open(path)

plt.figure(num='alignment', figsize = (8,8))

plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(im)
plt.axis('off')

im = np.asarray(im)

face_detector = dlib.get_frontal_face_detector()
face_locations = face_detector(im, upsample_num_times = 1)

predictor_5_point_model = "/Users/yujun/VirtualEnvs/face_recognition/lib/python3.5/site-packages" \
                            "/face_recognition_models/models/shape_predictor_5_face_landmarks.dat"
pose_predictor_point = dlib.shape_predictor(predictor_5_point_model)

faces = dlib.full_object_detections()
for face_location in face_locations:
    faces.append(pose_predictor_point(im, face_location))

im = dlib.load_rgb_image(path)
img = dlib.get_face_chip(im, faces[0])

plt.subplot(1,2,2)
plt.title('alignment image')
plt.imshow(img)
plt.axis('off')
'''
```
