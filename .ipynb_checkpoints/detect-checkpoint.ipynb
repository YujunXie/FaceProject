# 人脸检测

使用dlib库进行人脸检测。

### 导入图像

```python
import cv2
import numpy as np
import dlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

%matplotlib inline

path = 'Detection/data/Image/Ai_Sugiyama_0001.jpg'
im = Image.open(path)
im = np.array(im)

plt.axis('off')
plt.imshow(im)
```

### 使用基于提取hog特征的人脸检测器

This face detector is made using the now classic Histogram of Oriented Gradients (HOG) feature combined with a linear classifier(regression tree), an image pyramid, and sliding window detection scheme.

```python
#hog feature
face_detector = dlib.get_frontal_face_detector()
faces = face_detector(im, upsample_num_times = 1)

print("found {} face(s) in this photograph.".format(len(faces)))
```

```python
plt.figure()
plt.axis('off')
plt.imshow(im)
currentAxis = plt.gca()

cnt = 1
# Print the location of each face in this image
for face in faces:
    
    top, right, bottom, left = max(face.top(), 0), min(face.right(), im.shape[1]), min(face.bottom(), im.shape[0]), max(face.left(), 0)
    print("A face is located at pixel location Left: {}, Top: {}, Right: {}, Bottom: {}".format(left, top, right, bottom))
    
    rect = patches.Rectangle((left, top), right - left, bottom - top, linewidth=2, edgecolor='y', facecolor='none')
    currentAxis.add_patch(rect)
    
    box = (left, top, right, bottom)
    face_image = Image.open(path).crop(box)
    face_image.save('Detection/data/FaceImage/Ai_Sugiyama_0001_face.jpg'.format(cnt))
    cnt += 1
   
```

### 使用训练好的cnn提取特征的人脸检测器

需要用GPU加速，否则运行时间较长。

```python
#cnn feature
cnn_face_detection_model = "/Users/yujun/VirtualEnvs/face_recognition/lib/python3.5/site-packages" \
                            "/face_recognition_models/models/mmod_human_face_detector.dat"
cnn_face_detector = dlib.cnn_face_detection_model_v1(cnn_face_detection_model)

faces = cnn_face_detector(im, upsample_num_times = 1)

print("found {} face(s) in this photograph.".format(len(faces)))
```

```python
plt.figure()
plt.axis('off')
plt.imshow(im)
currentAxis = plt.gca()

for face in faces:
    
    top, right, bottom, left = max(face.rect.top(), 0), min(face.rect.right(), im.shape[1]), min(face.rect.bottom(), im.shape[0]), max(face.rect.left(), 0)
    print("A face is located at pixel location Left: {}, Top: {}, Right: {}, Bottom: {}".format(left, top, right, bottom))
    
    rect = patches.Rectangle((left, top), right - left, bottom - top, linewidth=2, edgecolor='r', facecolor='none')
    currentAxis.add_patch(rect)
```

# 应用-摄像头实时检测人脸

使用dlib库进行检测时，速度较慢，准确性更高。
使用opencv库进行检测时，速度快，准确性差一点。

```python
def discern(img):
    
    #dlib检测器
    '''
    faces = face_detector(img, upsample_num_times = 1)
    for face in faces:
        top, right, bottom, left = max(face.top(), 0), min(face.right(), img.shape[1]), min(face.bottom(), img.shape[0]), max(face.left(), 0)
        cv2.rectangle(img, (left, top, (right, bottom, (0, 255, 0), 2)
        cv2.imshow("image", img)
    '''
    
    #opencv检测器
    cap = cv2.CascadeClassifier("/Users/yujun/VirtualEnvs/face_recognition/lib/python3.5/site-packages/cv2/data" \
                               "/haarcascade_frontalface_default.xml")
    faceRects = cap.detectMultiScale(
        img, scaleFactor=1.2, minNeighbors=3, minSize=(50, 50))
    if len(faceRects):
        for faceRect in faceRects:
            x, y, w, h = faceRect
            cv2.rectangle(img, (x, y), (x + h, y + w), (0, 255, 0), 2)
    cv2.imshow("Image", img)
    
cap = cv2.VideoCapture(0)
while(1):
    ret, img = cap.read()
    discern(img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
